```python

from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
```

#sample text

```python
text="Word embeddings are essential in NLP taskes. They capture semantic relationship between words"
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences=sent_tokenize(text)
tokenized_sentences=[word_tokenize(sentence.lower()) for sentence in sentences]
```

```python
print(tokenized_sentences)
```

# Train the word2Vec Model

```python
print("Train the Word2Vec Model")
model=Word2Vec(tokenized_sentences,vector_size=100,window=5,min_count=1,sg=0)
print("Word2Vec model trained successfully")
print()
```

#Access the words vetors

```python
print("Access word vectors")
word_vectors=model.wv
vector=word_vectors['word']
print("Vectors for 'word:",vector)
print()
```

#finding similar words

```python
print("finding similar words")
similar_words=model.wv.most_similar('embeddings',topn=3)
print("Similar word to 'embedding':",similar_words)
```

#tokenize into words


```python
print("tokenize into words")
words=word_tokenize(text)
print("Tokenized words:",words)
print()
```

#Iniate the Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
print("Inialize the lemmatizer")
lemmatizer=WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

#Lemmatize words

```python
print("Lemmatize words")
lemmatized_words=[lemmatizer.lemmatize(word)for word in words]
print("Lemmatized words:", lemmatized_words)
```
